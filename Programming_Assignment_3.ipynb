```python
1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?
Ans:-
a=int(input('Enter a number to check if it is +ve, -ve or zero'))

if a<0:
    print('The number you have entered is Negative')
elif a>0:
    print('The number you have entered is Positive')
else:
    print('The number is 0')
```

```python
2.	Write a Python Program to Check if a Number is Odd or Even?
Ans:-
a=int(input('Enter a number to check if it is even or odd'))

if (a%2)==0:
    print('The number {} is even'.format(a))
else:
    print('The number {} is odd'.format(a))
```

```python
3.	Write a Python Program to Check Leap Year?
Ans:-
year=int(input('Enter a year to check if it is Leap Year or not'))

if (year%4)==0:
    print('The year {} is Leap Year'.format(year))
else:
    print('The year {} is not a Leap Year'.format(year))
```

```python
4.	Write a Python Program to Check Prime Number?
Ans:-
num=int(input('Enter a number to check if it is Prime Number or not'))
count=0
i=1
while(i<=num):
    if num%i==0:
        count=count+1
    i=i+1
if (count==2):
    print('The number {} is a Prime Number'.format(num))
else:
    print('The number {} is not a Prime Number'.format(num))
```
