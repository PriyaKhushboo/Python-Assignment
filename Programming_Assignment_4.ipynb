```python
4.	Write a Python Program to Check Armstrong Number?
Ans:-
    
i=int(input('Enter the number'))
num =i
result=0
n=len(str(i))
while(i !=0):
    digit=i%10
    result=result+digit**n
    i=i//10
if num==result:   
    print('Yes, This is an Armstrong number')
else:
    print('No, This is not an Armstrong number')
```

```python
5.	Write a Python Program to Find Armstrong Number in an Interval?
Ans:-suppose we want to find Armstrong number in an interval of 0 to 100
    
for i in range(101):
    num =i
    result=0
    n=len(str(i))
    while(i !=0):
        digit=i%10
        result=result+digit**n
        i=i//10
    if num==result:
        print(num)    
```
