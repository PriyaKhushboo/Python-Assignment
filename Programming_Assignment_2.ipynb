```python
1.	Write a Python program to convert kilometers to miles?
Ans:-
a=int(input('Enter the kilometers distance which you want to convert into miles: '))
res=a* 0.62137119
print('{} kilometer in miles is '.format(a),res)
```

```python
2.	Write a Python program to convert Celsius to Fahrenheit?
Ans:-
a=int(input('Enter the temperature in celcius to convert into Farenheit: '))
print('The temperature in Farenheit is ',(a*1.8)+32)    
```

```python
4.	Write a Python program to solve quadratic equation?
Ans:-
import cmath
a=int(input('enter the 1st number'))
b=int(input('enter the 2nd number'))
c=int(input('enter the 3rd number'))
d=(b**2) - (4*a*c)
sol_1=(-b+cmath.sqrt(d))/(2*a)
sol_2=(+b-cmath.sqrt(d))/(2*a)
print('The answers are {} and {} '.format(sol_1,sol_2))
```

```python
5.	Write a Python program to swap two variables without temp variable?
Ans:-
x=2
y=4
print('Before swapping: ',x,y)
x,y=y,x
print('After swapping: ',x,y)
```
