```python
17. How do you distinguish between copy.copy() and copy.deepcopy()?
Ans:-copy.copy is used for shallow copy and copy.deepcopy is for deep copy. when we try to perform these deep copy and shallow
    copy operation on single normal list then these two will be same but when we will try to perform these two operation on
    nested list then these two are different as in deep copy if we will change an element of a nested list it will not be
    copied in other list as it will create a different memory location but in shallow copy, both the list will get updated as 
    both the list points to the same memory location.
```
