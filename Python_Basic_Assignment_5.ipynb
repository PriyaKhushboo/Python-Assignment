```python
6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?
Ans:-The expression 'cat' in spam will look for if there is cat present anywhere in the dictionary as a key or value or not
    and 'cat' in spam.values() will look for if cat is present specifically as values or not.
```
