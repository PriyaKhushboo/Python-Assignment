```python
11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Ans:-These are 3 ways to call range function.
     range(10) means here it is considering 10 as the end point and whenever we call this we will get a series of number 
     starting from zero as default excluding 10. 
    
     range(0,10) means here it will consider 0 as the starting point and 10 as the end point and whenever we call this we 
     will get a series of number starting from zero till 9 excluding 10
    
     range(0,10,1) means here it will consider 0 as the starting point, 10 as the end point and 1 as the step and whenever
     we call this we will get a series of number starting from zero till 9 excluding 10 with a step of 1.
```
