```python
1.	Write a Python Program to Find LCM?
Ans:-
    
def compute_LCM(n1,n2):
    if n1>n2:
        higher=n1
    else:
        higher=n2
    a=higher
    while True:
        if higher%n1==0 and higher%n2==0:
            print('The LCM of n1 and n2 is:',higher)
            break
        else:
            higher=higher+a
            
n1=int(input('Enter the 1st number'))
n2=int(input('Enter the 2nd number'))
compute_LCM(n1,n2)
```

```python
2.	Write a Python Program to Find HCF?
Ans:-
    
def compute_HCF(n1,n2):
    if n2==0:
        return n1
    else:
        return compute_HCF(n2, n1%n2)
    
n1=int(input('Enter the 1st number'))
n2=int(input('Enter the 2nd number'))
print('The HCF of n1 and n2 is: ',compute_HCF(n1,n2))
```

```python
3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
Ans:-
    
n=int(input('Enter a number'))
print('The binary form of ', n, 'is', bin(n))
print('The binary form of ', n, 'is', oct(n))
print('The binary form of ', n, 'is', hex(n))
```

```python
4.	Write a Python Program To Find ASCII value of a character?
Ans:-
    
n=input('Enter the character')
for i in n:
    print('The ASCII value of ',n,'is', ord(i), end=':')
```

```python
5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?
Ans:-
    
a=float(input('Enter 1st number'))
b=float(input('Enter 2nd number'))
operation=input('Enter the operation which you want to perform')
if operation =='+':
    print(a+b)
elif operation =='-':
    print(a-b)
elif operation =='*':
    print(a*b)
else:
    print(a/b)
```
